```python
# #Once created the clone of GIT-HUB repository then,
# #Required libraries for the program
import pandas as pd
import json
import os

# #This is to direct the path to get the data as states

path1=r"D:\Swetha Documents\Python_Code\CAPSTONE_SEC_PHONEPE\files\data\aggregated\transaction\country\india\state/"
Agg_state_list=os.listdir(path1)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

# Aggregate Transaction

data_1={'State':[], 'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list:

    state_path=path1+i+"/"
    
    Agg_yr=os.listdir(state_path)

    for j in Agg_yr:

        year_path=state_path+j+"/"

        Agg_yr_list=os.listdir(year_path)

        for k in Agg_yr_list:

            year_list=year_path+k

            Data=open(year_list,'r')

            D=json.load(Data)

            for z in D['data']['transactionData']:

                Name=z['name']
                data_1['Transaction_type'].append(Name)

                count=z['paymentInstruments'][0]['count']
                data_1['Transaction_count'].append(count)

                amount=z['paymentInstruments'][0]['amount']
                data_1['Transaction_amount'].append(amount)

                data_1['State'].append(i)
                data_1['Year'].append(j)
                data_1['Quarter'].append(int(k.strip('.json')))

#Succesfully created a dataframe
Agg_Trans=pd.DataFrame(data_1)

Agg_Trans["State"] = Agg_Trans["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Agg_Trans["State"] = Agg_Trans["State"].str.replace("-"," ").str.title()
Agg_Trans['State'] = Agg_Trans['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
Agg_Trans
```

```python
import pandas as pd
import json
import os

# #This is to direct the path to get the data as states

path2=r"D:\Swetha Documents\Python_Code\CAPSTONE_SEC_PHONEPE\files\data\aggregated\user\country\india\state/"
Agg_users_list=os.listdir(path2)
Agg_users_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

# Aggregated User

data_2={'State':[], 'Year':[],'Quarter':[],'Brand':[],'Counts':[],'Percentage':[]}

for a in Agg_users_list:

    first_path=path2+a+"/"
    
    yr=os.listdir(first_path)

    for b in yr:

        second_path=first_path+b+"/"

        user_li=os.listdir(second_path)

        for c in user_li:

            last_file=second_path+c

            Data=open(last_file,'r')

            ag_user=json.load(Data)
            
            try:

                for u in ag_user['data']['usersByDevice']:

                    Brand=u['brand']
                    data_2['Brand'].append(Brand)

                    Count1=u['count']
                    data_2['Counts'].append(Count1)

                    Percentage=u['percentage']
                    data_2['Percentage'].append(Percentage)

                    data_2['State'].append(a)

                    data_2['Year'].append(b)

                    data_2['Quarter'].append(int(c.strip('.json')))
            
            except:
                pass

#Succesfully created a dataframe
Agg_Users=pd.DataFrame(data_2)

Agg_Users["State"] = Agg_Users["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Agg_Users["State"] = Agg_Users["State"].str.replace("-"," ").str.title()
Agg_Users['State'] = Agg_Users['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
Agg_Users
```

```python
import pandas as pd
import json
import os

# #This is to direct the path to get the data as states

path3=r"D:\Swetha Documents\Python_Code\CAPSTONE_SEC_PHONEPE\files\data\map\transaction\hover\country\india\state/"
map_transaction_path=os.listdir(path3)
map_transaction_path
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

# Map Transaction

data_3={'State':[], 'Year':[],'Quarter':[],'Districts':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in map_transaction_path:

    map_path=path3+i+"/"
    
    map_yr=os.listdir(map_path)

    for j in map_yr:

        year=map_path+j+"/"

        map_list=os.listdir(year)

        for k in map_list:

            files=year+k

            Data=open(files,'r')

            M=json.load(Data)

            for m in M['data']['hoverDataList']:

                Districts=m['name']
                data_3['Districts'].append(Districts)

                count=m['metric'][0]['count']
                data_3['Transaction_count'].append(count)

                amount=m['metric'][0]['amount']
                data_3['Transaction_amount'].append(amount)

                data_3['State'].append(i)
                data_3['Year'].append(j)
                data_3['Quarter'].append(int(k.strip('.json')))

#Succesfully created a dataframe
Map_Transaction=pd.DataFrame(data_3)

Map_Transaction["State"] = Map_Transaction["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Map_Transaction["State"] = Map_Transaction["State"].str.replace("-"," ").str.title()
Map_Transaction['State'] = Map_Transaction['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
Map_Transaction
```

```python
import pandas as pd
import json
import os

# #This is to direct the path to get the data as states

path4=r'D:\Swetha Documents\Python_Code\CAPSTONE_SEC_PHONEPE\files\data\map\user\hover\country\india\state/'
map_user_path=os.listdir(path4)
map_user_path

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

# Map User

data_4={'State':[], 'Year':[],'Quarter':[],'Districts':[], 'RegisteredUsers':[], 'AppOpens':[]}

for i in map_user_path:

    m_u=path4+i+"/"
    
    m_yr=os.listdir(m_u)

    for j in m_yr:

        map1=m_u+j+"/"

        map_yr_list=os.listdir(map1)

        for k in map_yr_list:

            file=map1+k

            Datas=open(file,'r')

            map_user=json.load(Datas)
            
            for b in map_user['data']['hoverData'].items():

                District = b[0]
                data_4['Districts'].append(District)
                
                RegisteredUsers = b[1]['registeredUsers']
                data_4['RegisteredUsers'].append(RegisteredUsers)
                
                AppOpens = b[1]['appOpens']
                data_4['AppOpens'].append(AppOpens)

                data_4['State'].append(i)
                data_4['Year'].append(j)
                data_4['Quarter'].append(int(k.strip('.json')))

#Succesfully created a dataframe
Map_User=pd.DataFrame(data_4)

Map_User["State"] = Map_User["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Map_User["State"] = Map_User["State"].str.replace("-"," ").str.title()
Map_User['State'] = Map_User['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
Map_User
```

```python
import pandas as pd
import json
import os

# #This is to direct the path to get the data as states

path5=r"D:\Swetha Documents\Python_Code\CAPSTONE_SEC_PHONEPE\files\data\top\transaction\country\india\state/"
top_transaction=os.listdir(path5)
top_transaction
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

# Top Transaction

data_5={'State':[], 'Year':[],'Quarter':[],'Pincodes':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in top_transaction:

    app=path5+i+"/"
    
    map_yr=os.listdir(app)

    for j in map_yr:

        lin=app+j+"/"

        map_list=os.listdir(lin)

        for k in map_list:

            lin2=lin+k

            Data=open(lin2,'r')

            T=json.load(Data)
            
            for t in T["data"]["pincodes"]:
                
                Pincode=t['entityName']
                data_5['Pincodes'].append(Pincode)

                count=t['metric']['count']
                data_5['Transaction_count'].append(count)

                amount=t['metric']['amount']
                data_5['Transaction_amount'].append(amount)

                data_5['State'].append(i)
                data_5['Year'].append(j)
                data_5['Quarter'].append(int(k.strip('.json')))

#Succesfully created a dataframe
Top_Transaction=pd.DataFrame(data_5)

Top_Transaction["State"] = Top_Transaction["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Top_Transaction["State"] = Top_Transaction["State"].str.replace("-"," ").str.title()
Top_Transaction['State'] = Top_Transaction['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
Top_Transaction
```

```python
import pandas as pd
import json
import os

# #This is to direct the path to get the data as states

path6=r"D:\Swetha Documents\Python_Code\CAPSTONE_SEC_PHONEPE\files\data\top\user\country\india\state/"
top_user_path=os.listdir(path6)
top_user_path

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

# Top Users

data_6={'State':[], 'Year':[],'Quarter':[],'Pincodes':[], 'RegisteredUsers':[]}

for i in top_user_path:

    app=path6+i+"/"
    
    map_yr=os.listdir(app)

    for j in map_yr:

        lin=app+j+"/"

        map_list=os.listdir(lin)

        for k in map_list:

            lin2=lin+k

            Data=open(lin2,'r')

            T=json.load(Data)
            
            for top_user in T["data"]["pincodes"]:
                
                Pincodes=top_user['name']
                data_6['Pincodes'].append(Pincodes)

                Reg_Users=top_user['registeredUsers']
                data_6['RegisteredUsers'].append(Reg_Users)

                data_6['State'].append(i)
                data_6['Year'].append(j)
                data_6['Quarter'].append(int(k.strip('.json')))

#Succesfully created a dataframe
Top_User=pd.DataFrame(data_6)

Top_User["State"] = Top_User["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Top_User["State"] = Top_User["State"].str.replace("-"," ").str.title()
Top_User['State'] = Top_User['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
Top_User
```
